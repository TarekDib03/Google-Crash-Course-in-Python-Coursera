##### Reading and Processing Text Data

```python
dream_speech = open("I have a Dream - MLK.txt", "r", encoding='utf-8')
def calculate_frequencies(file_contents):
    # Here is a list of punctuations and uninteresting words we may use to clean the file content
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", 
                           "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", 
                           "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been",
                           "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", 
                           "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", 
                           "will", "just", "in", "for", "not"]
    
    # Remove punctuations
    cleaned_text = ""
    for char in file_contents.read().lower():
        if char not in punctuations:
            cleaned_text = cleaned_text + char
    cleaned_text = cleaned_text.split()

    # Remove uninteresting words
    for uninteresting_word in uninteresting_words:
        while uninteresting_word in cleaned_text:
            cleaned_text.remove(uninteresting_word)

    # Count the nummber of occurences of words
    word_freq = {}
    for word in cleaned_text:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1
            
    # wordcloud
    cloud = wordcloud.WordCloud(width = 2000, height = 1200)
    cloud.generate_from_frequencies(word_freq)
    return cloud.to_array()
```

##### Word Cloud

```python
myimage = calculate_frequencies(dream_speech)
plt.figure(figsize=(20,10), facecolor='k')
plt.imshow(myimage)
plt.axis("off")
plt.savefig('MLK_Dream_Speech.png', facecolor='k', bbox_inches='tight')
```
